```python
# Importing modules
import matplotlib.pyplot as mp
import numpy as np
```

```python
#sine function
x=np.arange(0,10,0.25)
y=np.sin(x)
mp.plot(x,y)
```

```python
#logarithm function
x=np.arange(0.1,10,0.1)
y=np.log10(x)
mp.plot(x,y)
```

```python
# y=x^2 -4x +3
x=np.arange(-3,7,0.25)
y=[i*i-4*i+3 for i in x]
mp.plot(x,y)
```
